```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load csv files
customers = pd.read_csv('Customers.csv')
products = pd.read_csv('Products.csv')
transactions = pd.read_csv('Transactions.csv')
```

# Insights From Customers.csv

```python
print(customers.info())
```

```python
print(customers.describe())
```

```python
# plot customers from each region (continent)
region_counts = customers['Region'].value_counts()
region_counts.plot(kind='bar', figsize=(8, 6), color='skyblue')
plt.title('Number of Customers by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
# plot Monthly Customer Signups
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])
customers['SignupMonth'] = customers['SignupDate'].dt.to_period('M')
signup_trend = customers['SignupMonth'].value_counts().sort_index()
signup_trend.plot(kind='line', figsize=(10, 6), marker='o')
plt.title('Monthly Customer Signups')
plt.xlabel('Month')
plt.ylabel('Number of Signups')
plt.show()
```

# Insights From Products.csv

```python
print(products.info())
```

```python
print(products.describe())
```

```python
# most expensive products
top_expensive_products = products.sort_values(by='Price', ascending=False).head(10)
print("Top 10 Most Expensive Products:")
print(top_expensive_products[['ProductID', 'ProductName', 'Price']])
```

```python
# plot price distribution of products
sns.histplot(data=products, x='Price', bins=20, kde=True)
plt.title('Price Distribution of Products')
plt.show()
```

```python
# creating boxplot for price distribution by category of product
sns.boxplot(data=products, x='Category', y='Price', hue='Category', palette='Set2')
plt.title('Price Distribution by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
product_aggregates = transactions.groupby('ProductID').agg({'Quantity': 'sum'}).reset_index()
top_products = product_aggregates.sort_values(by='Quantity', ascending=False).head(10)
top_products = top_products.merge(products[['ProductID', 'ProductName']], on='ProductID', how='left')
print(top_products)

# Plot the top 10 highest purchased products
plt.figure(figsize=(12, 6))
plt.bar(top_products['ProductName'], top_products['Quantity'])
plt.title('Top 10 Highest Purchased Products')
plt.xlabel('Product Name')
plt.ylabel('Total Quantity Purchased')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Insights From Transactions.csv

```python
print(transactions.info())
```

```python
print(transactions.describe())
```

```python
# plotting monthly sales trend
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])
transactions['TransactionMonth'] = transactions['TransactionDate'].dt.to_period('M')
sales_trend = transactions.groupby('TransactionMonth')['TotalValue'].sum()
sales_trend.plot(kind='line', figsize=(10, 6), marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales (USD)')
plt.show()
```

```python
# plot sales by region
customer_region_sales = pd.merge(transactions, customers, on='CustomerID')
region_sales = customer_region_sales.groupby('Region')['TotalValue'].sum().sort_values()
region_sales.plot(kind='bar', figsize=(8, 6), color='green')
plt.title('Total Sales by Region')
plt.ylabel('Total Sales (USD)')
plt.xlabel('Region')
plt.show()
```

```python
# Merge transactions, customers, and products data
customer_region_sales = pd.merge(transactions, customers, on='CustomerID')
category_region_sales = pd.merge(customer_region_sales, products, on='ProductID')

category_region_group = category_region_sales.groupby(['Category', 'Region'])['Quantity'].sum().reset_index()
pivot_table = category_region_group.pivot(index='Category', columns='Region', values='Quantity')

# Plot: Categories vs Regions
plt.figure(figsize=(12, 8))
pivot_table.plot(kind='bar', stacked=False, figsize=(10, 6), colormap='tab10')
plt.title('Quantity Sold by Region for Each Category')
plt.ylabel('Quantity Sold')
plt.xlabel('Product Category')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Aggregate data by CustomerID
customer_aggregates = transactions.groupby('CustomerID').agg({'Quantity': 'sum', 'TotalValue': 'sum'}).reset_index()
# Sort by Quantity
top_customers_by_quantity = customer_aggregates.sort_values(by='Quantity', ascending=False)
# Merge the customer aggregates with the customer data to get region information
customer_aggregates_with_region = customer_aggregates.merge(customers[['CustomerID', 'Region']], on='CustomerID', how='left')

# Plot top customers by quantity with region information
plt.figure(figsize=(12, 6))
sns.barplot(x='CustomerID', y='Quantity', hue='Region', data=customer_aggregates_with_region.sort_values(by='Quantity', ascending=False).head(10))
plt.title('Top 10 Customers by Quantity Purchased with Region')
plt.xlabel('Customer ID')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.show()
```

```python
# Sort by total sales
top_customers_by_total_value = customer_aggregates.sort_values(by='TotalValue', ascending=False)
# Plot top customers by total sales with region information
plt.figure(figsize=(12, 6))
sns.barplot(x='CustomerID', y='TotalValue', hue='Region', data=customer_aggregates_with_region.sort_values(by='TotalValue', ascending=False).head(10), palette='magma')
plt.title('Top 10 Customers by Total Sales with Region')
plt.xlabel('Customer ID')
plt.ylabel('Total Sales (USD)')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.show()
```
